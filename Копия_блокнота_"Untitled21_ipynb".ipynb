```python
ar = np.array([[[random.randint(0,101) for i in range(0,3)] for j in range(0,3)] for k in range(0,3)])
print(ar)
```

```python
size = random.randint(3, 8)
ar = np.array([[(1 if i % size == 0 or i == size - 1 or j == size - 1 or j % size == 0 else 0) for i in range(0, size)] for j in range(0, size)])
print(ar)
```

```python
ar = np.array([random.randint(0, 101) for i in range(0, 10)])
ar = np.sort(ar)
ar = ar[::-1]
print(ar)
```

```python
dot1 = tuple([random.randint(-100, 100)/10 for i in range(0,3)])
dot2 = tuple([random.randint(-100, 100)/10 for i in range(0,3)])
print(dot1, dot2)
r = math.sqrt((dot1[0] - dot2[0]) ** 2 + (dot1[1] - dot2[1]) ** 2 + (dot1[2] - dot2[2]) ** 2)
print(r)
```
